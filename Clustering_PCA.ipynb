# Secon example:

```python
# importing the libraires:
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
# loading the dataset:
df = pd.read_csv('heightweight.csv')
df
```

```python
df['gender'].value_counts()
```

```python
# encoding the gender to 0 and 1:
df['gender'] = df.gender.apply(lambda x: 1 if x =='f' else 0 )
df
```

```python
# setting the PCA number of components:
pca = PCA(n_components=2)
```

```python
df.shape
```

```python
# creating a new dataframe with the PCA data:
new_df = pd.DataFrame(pca.fit_transform(df), columns=['PCA1', 'PCA2'])
new_df
```

```python
# variance ratio:
pca.explained_variance_ratio_.sum()
```

```python
# plotting the PCA:
plt.scatter(new_df.PCA1, new_df.PCA2)
```

```python
# initializing the kmeans model:
model = KMeans(n_clusters=3)
model.fit_transform(new_df)
```

```python
model.labels_
```

```python
pd.Series(model.labels_).value_counts()
```

```python
# getting the cluster centers:
model.cluster_centers_
```

```python
# plotting the clusters before scaling:
plt.scatter(new_df.PCA1, new_df.PCA2, c = model.labels_)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], marker="X", c='r')
```

```python
# Scaling the data:
scaler = MinMaxScaler()
x_scaled= pd.DataFrame(scaler.fit_transform(new_df), columns=new_df.columns)
x_scaled
```

```python
# initializing the second kmeans model:
model_v2 = KMeans(n_clusters=3)
model_v2.fit_transform(x_scaled)
model_v2.labels_
```

```python
# plotting the cluster after scaling:
plt.scatter(x_scaled.PCA1, x_scaled.PCA2, c = model_v2.labels_)
plt.scatter(model_v2.cluster_centers_[:,0], model_v2.cluster_centers_[:,1], marker="X", c='r')
```
