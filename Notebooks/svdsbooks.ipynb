```python
um = df.pivot_table(index='User-ID', columns=['Book-Title'], values='Book-Rating', fill_value=0)
um
```

# Constructing the svd

```python
A = um.values
```

```python
new_um = df.pivot_table(index='User-ID', columns=['Book-Title'], values='Book-Rating')
new_um
```

```python
um.iloc[0]
```

```python
from scipy.sparse.linalg import svds
```

```python
U, sigma, vt = svds(A, k=100)
U, sigma, vt
```

```python
pd.DataFrame(U)
```

```python
pd.DataFrame(vt)
```

```python
sigma
```

```python
import numpy as np
```

```python
s = np.diag(sigma)
s
```

# re constructing the svd

```python
r_um =np.dot((np.dot(U,s)),vt)
```

```python
rum_df = pd.DataFrame(r_um, columns=um.columns)
rum_df
```

```python
rum_df.iloc[0].sort_values(ascending=False)
```

```python
movie_watched_user1 = new_um.iloc[0].dropna().index
movie_watched_user1
```
