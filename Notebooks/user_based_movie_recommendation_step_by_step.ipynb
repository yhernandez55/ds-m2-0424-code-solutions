# read data

```python
import pandas as pd
```

```python
movie = pd.read_csv('u.item', encoding='latin1', delimiter='|', header=None, usecols=[0,1,2], names=['movie_id', 'name', 'date'])
movie
```

```python
rating = pd.read_csv('u.data', header=None, delimiter='\t', names=['user_id','movie_id', 'ratings', 'time'])
rating
# time: epoc time or unex time 
```

```python
u_user = pd.read_csv('u.user', delimiter='|', header=None)
u_user
```

# df

```python
df = rating.merge(movie, on=['movie_id'], how='inner')
df
```

```python
df[df.name=='Toy Story (1995)']
```

```python
df.groupby('name')['ratings'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('name').agg({'ratings': ['mean', 'count', 'sum']}).sort_values(by=('ratings', 'count'), ascending=False).head(10)
```

# um

```python
um = df.pivot_table(index='user_id', columns=['name'], values='ratings')
um
```

```python
popularmovies = df.groupby('name').agg({'ratings': ['mean', 'count', 'sum']}).sort_values(by=('ratings', 'count'), ascending=False)
popularmovies
```

```python
popularmovies = popularmovies[popularmovies['ratings']['count'] >= 100]
popularmovies
```

```python
similarmovies.shape, popularmovies.shape
```

```python
popularmovies = pd.DataFrame(popularmovies)
popularmovies
```

```python
popularmovies = pd.DataFrame(popularmovies).reset_index()
popularmovies.columns = ['name', 'mean', 'count', 'sum']
popularmovies
```

```python
similarmovies = pd.DataFrame(similarmovies).reset_index()
similarmovies
# collab filtering
```

```python
df = popularmovies.merge(similarmovies, on='name', how='inner')
df
# wanted to filter similar movies and just keep the popular ones and wanted similarity 
```

```python
df.drop(['mean', 'sum', 'count'], axis=1, inplace=True)
df
```

```python
df.rename(columns={'Star Wars (1977)': 'Similarity'}, inplace=True)
```

```python
df.set_index('name', inplace=True)
df
```

```python
df.sort_values(by = 'Similarity', ascending=False).head(20)
```
