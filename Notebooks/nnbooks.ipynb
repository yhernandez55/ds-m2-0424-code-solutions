```python
df = df.iloc[:10000]
```

```python
um = df.pivot_table(index='User-ID', columns=['Book-Title'], values='Book-Rating', fill_value=0)
um
```

```python
import numpy as np
from scipy import sparse
```

```python
print(sparse.csr_matrix(um.values))
```

```python
from sklearn.neighbors import NearestNeighbors
neighbor = NearestNeighbors(n_neighbors=5)
neighbor.fit(um)
```

```python
distance, user = neighbor.kneighbors(um,return_distance=True)
```
